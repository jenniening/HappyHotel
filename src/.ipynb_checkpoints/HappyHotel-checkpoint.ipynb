####  Common Word in Reviews

To get the word vocabulary, we need to do some simple preprocessing:
1. Split setence into words using gensim.utils.simple_preprocess
2. Remove punctuations, we determined based on gensim.parsing.preprocessing.STOPWORDS
3. Remove grammer effect on words: change third person into first, verbs into present form, using WordNetLemmatizer()
4. Remove uncommon words and too common words: here we remove words only appear less than 50 reviews, and words appear more than 50% of reviews. 

```python
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            result.append(WordNetLemmatizer().lemmatize(token, pos='v'))
    return result
```

```python
text = data["Description"][0:1].values[0]
print("Input review:", text)
print("Output words:", ",".join(preprocess(text)))
```

Based on this preprocessing, we can convert all reviews into word list. In next step, we create word_list, which includes all words for each review

```python
word_list = []
for idx, i in enumerate(list(data["Description"])):
    word_list.append(preprocess(i))
    if idx % 1000 == 0:
        print(idx)
```

After get the word_list, we used gensim.corpora.Dictionary to build the dictionary for all words. This dictionary contains a lot of information of words:
1. looping the dictionary (iteritems(), token2id, id2token), we can get the tocken id and corresponding word
2. dictionary.cfs --> token_id, how many instances of this token are contained in the documents.
3. dictionary.dfs --> token_id, how many documents contain this token.
4. dictionary.num_docs --> number of documents processed
5. dictionary.num_pos --> number of processed words

```python
dictionary = gensim.corpora.Dictionary(word_list)
### fiter the words too uncommon and too common
print("Number of words:", len(dictionary))
dictionary.filter_extremes(no_below=15, no_above=0.50)
print("Number of words after filtering:", len(dictionary))
```

Based on current dictionary, we can get the count information for different type of reviews

```python
### Get the dictionary of word count for different type of reviews
happy_dict = {key:0 for key in dictionary.token2id.keys()}
no_happy_dict = {key:0 for key in dictionary.token2id.keys()}
for i in word_list[0:26521]:
    for ii in i:
        if ii in happy_dict:
            happy_dict[ii] += 1
for i in word_list[26521:]:
    for ii in i:
        if ii in no_happy_dict:
            no_happy_dict[ii] += 1
```

### Topic Model Development

#### Get Bag of Words

We use dictionary.doc2bow to get the dictionary reporting **appear words** and **the number of times** in each review (document)

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in word_list]
```

```python
### example
print("Fist review:", data["Description"].values[0])
for i in bow_corpus[0]:
    print(dictionary[i[0]], "appear", i[1], "times")
```

#### LDA

We are going to use LDA model to generate topics from our reviews. However, one challenge is that we need to determine the number of topics before we build the model. To solve this problem, we are doing Grid Search based on change of coherence.

##### Baseline

We randomly choose num_topics as 10 with chunksize as 1000, and train the corpos 10 times. Consider this as our baseline model

```python
lda = LdaMulticore(corpus=bow_corpus, id2word=dictionary, num_topics=10, random_state=100,
                chunksize=1000, passes=10, workers=4)
```

```python
lda.show_topics()
```

```python
coherence_model_lda = CoherenceModel(model=lda, texts=word_list, dictionary=dictionary, coherence="c_v")
```

```python
coherence_lda = coherence_model_lda.get_coherence()
```

```python
print("Coherence of our baseline model", coherence_lda)
```

##### Grid Search

Since we need to determine the number of topics in LDA model, grid search has been conducted, and coherence for different number of topics have been recorded.

```python
def compute_coherence_values(corpus, dictionary, texts, k):
    lda_model = LdaMulticore(corpus=corpus, id2word=dictionary, num_topics=k, random_state=0, chunksize=1000, passes=10, per_word_topics=True, workers=4)
    
    coherence_model_lda = CoherenceModel(model=lda_model, texts=texts, dictionary=dictionary, coherence='c_v')
    
    return coherence_model_lda.get_coherence()
```

```python
# Topic range
min_topics = 2
max_topics = 15
step_size = 1
topics_range = range(min_topics, max_topics, step_size)
# Validation sets
num_of_docs = len(bow_corpus)
corpus_sets = [gensim.utils.ClippedCorpus(bow_corpus, int(num_of_docs*0.80)), 
               bow_corpus]
corpus_title = ['80% Corpus', '100% Corpus']
model_results = {'Validation_Set': [],
                 'Topics': [],
                 'Coherence': []
                }
# iterate through validation corpuses
for i in range(1, len(corpus_sets)):
    # iterate through number of topics
    for k in topics_range:
        cv = compute_coherence_values(corpus=corpus_sets[i], dictionary=dictionary, texts=word_list, k=k)
        # Save the model results
        model_results['Validation_Set'].append(corpus_title[i])
        model_results['Topics'].append(k)
        model_results['Coherence'].append(cv)
#pd.DataFrame(model_results).to_csv('lda_tuning_results.csv', index=False)
```

```python
ax = sn.scatterplot(model_results["Topics"], model_results["Coherence"])
plt.xticks(size=12)
plt.yticks(size=12)
plt.xlabel("Number of Topics", fontsize=14)
plt.ylabel("Coherence", fontsize=14)
plt.savefig("../data/figures/LDA_GridSearch.png", dpi=100);
```

Thus, best coherence has been achieved by number of topics = 13.

##### Determine score for hotel based on topic

To determine the score of hotel on each topic, we just use a very naive way: count the number of that topic appearing in reviews and divide it by the total number of reviews. Here, we use total reviews, only happy reviews, and only not happy review to get the score for hotel.

```python
def get_score(index, review_topic):
    total_review = len(index)
    topic_total = [review_topic[i] for i in index]
    topic_dict = {key: 0 for key in range(13)}
    for key in range(13):
        for review_topic in topic_total:
            for topic in review_topic:
                if topic[0] == key:
                    topic_dict[key] += 1
    score_dict = {key: topic_dict[key]/total_review for key in range(13)}
    return score_dict
```

```python
score_dic = {}
for i in range(1,11):
    index = list(data[(data["hotel_ID"] == i)].index)
    score = get_score(index, review_topic)
    happy_index = list(data[(data["hotel_ID"] == i)&(data["Is_Response"] == "happy")].index)
    happy_score = get_score(happy_index, review_topic)
    nohappy_index = list(data[(data["hotel_ID"] == i)&(data["Is_Response"] == "not happy")].index)
    nohappy_score = get_score(nohappy_index, review_topic)
    score_dic[i] = [score, happy_score, nohappy_score]
```
